```python
pop = pd.read_excel('data/gapminder_population.xlsx',index_col=0)
```

```python
pop.index.name = 'country'
pop = pop.reset_index()
pop
```

```python
pop_lon = pop.melt(id_vars='country', var_name='year', value_name='Total population')
pop_lon
```

```python
fert_life = fert_lon.merge(life_lon,how='outer')
fert_life
```

```python
fert_life_pop = fert_life.merge(pop_lon,how='outer')
fert_life_pop
```

```python
flpnona = fert_life_pop.dropna()
flpnona
```

```python
flpnona.reset_index(inplace=True)
flpnona
```

```python
flpnona=flpnona.drop('index',axis=1)
flpnona
```

```python
conti = pd.read_csv('data/continents.csv', delimiter = ';', index_col=0)
conti = conti.reset_index()
conti
```

```python
contilist = conti.continent.unique()
contilist
```

```python
fulltab = flpnona.merge(conti, on = "country" , how='left' )
fulltab
```

```python
fulltab
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import imageio
```

```python
plt.style.use('seaborn-dark')
sns.set_style('whitegrid')
sns.color_palette('tab10')
```

```python
images = []
for i in range(1960,2016) :
    fert_life_pop_subset = fulltab.loc[fulltab['year'] == i]
    sns.scatterplot(x='life_expectancy', y='fertility_rate' ,palette='tab10', hue = 'continent' ,size='Total population' , data=fert_life_pop_subset, alpha=0.75)
    plt.axis((10,90,0,9))
    plt.legend(bbox_to_anchor=(0, 1), loc=2, labels = contilist)
    filename = 'images/jupyter-{0}.png'.format(i)
    plt.savefig(filename,dpi=200)
    images.append(imageio.imread(filename))
    plt.close()

imageio.mimsave('images/output-jup2.gif', images, fps=20)
```
